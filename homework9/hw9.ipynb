Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the following:

```python
# define function:
def plot_functions():
    #takes two functions and plots their graphs while also returning a list of the values of the range
    #x-value domain:
    x = np.linspace(0, 10, 100)
    f_x = x**2
    g_x = np.sqrt(x)  #functions to be plotted
    #make plot
    plt.figure(figsize = (10,8))
    plt.plot(x, f_x, label=r'$f(x) = x^2$', color='blue', linestyle='-')
    plt.plot(x, g_x, label=r'$g(x) = \sqrt{x}$', color='green', linestyle='--')
    plt.xlabel('x-axis')
    plt.ylabel('f(x), g(x)')
    plt.title('Plot of $f(x) = x^2$ and $g(x) = sqrt{(x)}$')
    plt.legend()
    plt.show()
    f_values = f_x.tolist() #function values as lists
    g_values = g_x.tolist()
    return f_values, g_values
print(plot_functions())



    
    
```
